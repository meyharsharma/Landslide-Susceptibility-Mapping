# Importing files

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
landslide=pd.read_excel('Landslide_Coordinates.xlsx')
ls=landslide.drop(['Latitude','Longitude'],axis=1)
ls = ls.set_index('Sr. No.')
ls.head()
```

```python
nonlandslide=pd.read_excel('Non-Landslide_Coordinates.xlsx')
nls=nonlandslide.drop(['Latitude','Longitude'],axis=1)
nls = nls.set_index('Sr. No.')
nls.head()
```

# Encoding

```python
ls.Lithology=pd.Categorical(ls.Lithology)
ls.Land_Cover=pd.Categorical(ls.Land_Cover)
ls.Watershed=pd.Categorical(ls.Watershed)
```

```python
l=pd.get_dummies(ls,columns=['Lithology','Land_Cover','Watershed'])
l.head(20)
```

```python
nls.Lithology=pd.Categorical(nls.Lithology)
nls.Land_Cover=pd.Categorical(nls.Land_Cover)
nls.Watershed=pd.Categorical(nls.Watershed)
nl=pd.get_dummies(nls,columns=['Lithology','Land_Cover','Watershed'])
nl.head()
```

Splitting the data

```python
train_l = l.sample(frac = 0.8, random_state = 10,replace=False,axis=0)
train_nl = nl.sample(frac = 0.8, random_state = 10,replace=False,axis=0)
train_nl.head()
```

 Final train

```python
train_all = pd.concat([train_l,train_nl], axis=0).fillna(0)
train_all.Land_Cover_20= train_all.Land_Cover_20.astype(np.uint8)
train_all.head()
```

```python
cols = ['Distance from Shillong','Distance from Nearest River','Distance from Nearest Fault','Elevation','Slope','Aspect','VARI','Lithology_Gneissic_Complex','Lithology_Granite_Pluton','Lithology_Shillong_Group','Land_Cover_1','Land_Cover_2','Land_Cover_9', 'Land_Cover_11','Land_Cover_12',
        'Land_Cover_13','Land_Cover_20','Land_Cover_22','Watershed_1','Watershed_2','Watershed_3','Watershed_4','Watershed_5','Watershed_6','Watershed_7','Watershed_8','Landslide']
train_all = train_all.reindex(columns=cols)
train_all.head()
```

```python
subset_1=train_l.index
test_l=l.drop(subset_1)
subset_2=train_nl.index
test_nl=nl.drop(subset_2)
```

Final test

```python
# Saving the columns in a list
test_all = pd.concat([test_l,test_nl], axis=0).reindex(columns=cols).fillna(0)
test_all.Land_Cover_20= test_all.Land_Cover_20.astype(np.uint8)
test_all.head()
```

# Random Forest

Base Model

```python
x_train = train_all.drop('Landslide',axis=1)
y_train = train_all['Landslide']
```

```python
x_test = test_all.drop('Landslide',axis=1)
y_test = test_all['Landslide']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200, random_state= 10)
rfc.fit(x_train,y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```

# Checking variable importance

```python
# Saving feature names for later use
feature_list = list(x_train.columns)
# Get numerical feature importances
importances = list(rfc.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
# Set the style
plt.style.use('fivethirtyeight')
# list of x locations for plotting
x_values = list(range(len(importances)))
# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')
# Tick labels for x axis
plt.xticks(x_values, feature_list, rotation='vertical')
# Axis labels and title
plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
```

# Improving the model

Randomized Search

```python
from pprint import pprint
# Look at parameters used by our current forest
print('Parameters currently in use:\n')
pprint(rfc.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 5)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 90, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rfc_random = RandomizedSearchCV(estimator = rfc, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=10, n_jobs = -1)
# Fit the random search model
rfc_random.fit(x_train, y_train)
```

```python
rfc_random.best_params_
```

# XGBoosting

```python
# check xgboost version
from numpy import mean
from numpy import std
import xgboost
from xgboost import XGBRFClassifier
from sklearn.metrics import precision_recall_fscore_support as score
# Hyperparameter tuning
from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV, RandomizedSearchCV
from hyperopt import tpe, STATUS_OK, Trials, hp, fmin, STATUS_OK, space_eval
```

```python
# define the model
boosted_model = XGBRFClassifier(n_estimators=400, subsample=1, colsample_bynode=0.2,random_state=10)
boosted_model.get_params()
```

```python
# fit the model on the whole dataset
boosted_model.fit(x_train, y_train)
```

```python
# define the model evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=10)
# evaluate the model and collect the scores
n_scores = cross_val_score(boosted_model, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance, training accuracy
print('Mean Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
boosted_model_pred= boosted_model.predict(x_test)
print(confusion_matrix(y_test,boosted_model_pred))
print(classification_report(y_test,boosted_model_pred))
```

```python
#grid search
# Define the search space
param_grid = { 
    # Percentage of columns to be randomly samples for each tree.
    "colsample_bytree": [ 0.1,0.3, 0.5 , 0.8,1],
    # reg_alpha provides l1 regularization to the weight, higher values result in more conservative models
    "reg_alpha": [0, 0.5, 1, 5],
    # reg_lambda provides l2 regularization to the weight, higher values result in more conservative models
    "reg_lambda": [0, 0.5, 1, 5]
    }
# Set up score
scoring = ['recall']
# Set up the k-fold cross-validation
kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=10)
```

hyperparameter tuning

```python
# Define grid search
grid_search = GridSearchCV(estimator=boosted_model, 
                           param_grid=param_grid, 
                           scoring=scoring, 
                           refit='recall', 
                           n_jobs=-1, 
                           cv=kfold, 
                           verbose=0)
# Fit grid search
grid_result = grid_search.fit(x_train, y_train)
# Print grid search summary
grid_result
# Print the best score and the corresponding hyperparameters
print(f'The best score is {grid_result.best_score_:.4f}')
print('The best score standard deviation is', round(grid_result.cv_results_['std_test_recall'][grid_result.best_index_], 4))
print(f'The best hyperparameters are {grid_result.best_params_}')
```

Fitting the model

```python
boosted_model1 = XGBRFClassifier(n_estimators=500, subsample=1, 
                                colsample_bytree=1,reg_alpha=0,reg_lambda=0.5,colsample_bynode=0.2,random_state=10)
```

```python
boosted_model1.fit(x_train, y_train)
```

```python
# define the model evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=10)
# evaluate the model and collect the scores
n_scores = cross_val_score(boosted_model1, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance, training accuracy
print('Mean Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
boosted_model1_pred= boosted_model1.predict(x_test)
print(confusion_matrix(y_test,boosted_model1_pred))
print(classification_report(y_test,boosted_model1_pred))
```
